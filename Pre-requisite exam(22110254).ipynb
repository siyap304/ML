# Question 1
## How many multiplications and additions do you need to perform a matrix multiplication between a (n, k) and (k, m) matrix? Explain.

Let matrix A be of the size (n,k) and matrix B of the size (k,m). So, the resulting matrix C will be of the size (n,m). Each element C[i, j] in the resulting matrix is the dot product of matrix A's i-th row and matrix B's j-th column.
To compute a single element C[i, j] k multiplications (for the dot product) and k-1 additions (to sum up the results of the multiplication) will be necessary.

Since the resulting matrix C has dimensions (n,m), the number of elements to be computed is n * m. Therefore, the total number of multiplications and additions for the entire matrix multiplication is given by:

Total multiplication = n * m * k

Total additions = n * m * (k-1)

So, for a matrix multiplication between a (n, k) matrix and a (k, m) needs n * m * k multiplications and n * m * (k-1)  additions.

# Question 2
## Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy. Compare the timing of both solutions. Which one is faster? Why?

```python
import time
start_time_lists = time.time()

def multiply_lists(a, b):
    c = [[0 for _ in range(len(b[0]))] for _ in range(len(a))]
    
    for i in range(len(a)):
        for j in range(len(b[0])):
            for k in range(len(b)):
                c[i][j] += a[i][k] * b[k][j]
    
    return c

matrix_a = [[11, 13, 15], [17, 19, 11], [13, 15, 17]]
matrix_b = [[12, 14, 16], [18, 12, 14], [16, 18, 12]]

matrix_c = multiply_lists(matrix_a, matrix_b)

end_time_lists = time.time()
print("Matrix after multiplication in lists:")
for row in matrix_c:
    print(row)

print("Time Taken in multiplication by lists:", (end_time_lists - start_time_lists))
```

```python
import time
import numpy as np

start_time_numpy = time.time()

matrix_a_np = np.array(matrix_a)
matrix_b_np = np.array(matrix_b)

matrix_c = np.dot(matrix_a, matrix_b)

end_time_numpy = time.time()

print("Matrix after multiplication in numpy:")
print(matrix_c)

print("Time Taken in multiplication by numpy:", (end_time_numpy - start_time_numpy))
```

## Conclusion
Here we can observe that the time taken by the second code i.e. the code which uses numpy is less than lists. This is because a numpy array is a collection of homogeneous data-types stored in contiguous memory locations. On the other hand, a list in Python is a collection of heterogeneous data types stored in non-contiguous memory locations. Moreover, the NumPy package breaks down a task into multiple fragments and then processes all the fragments parallelly.

# Question 3
## Finding the highest element in a list requires one pass of the array. Finding the second highest element requires 2 passes of the the array. Using this method, what is the time complexity of finding the median of the array? Can you suggest a better method? Can you implement both these methods in Python and compare against numpy.median routine in terms of time? 

If the highest element in the list requires one pass of the array and the second highest requires 2 passes of the array, then by this method, we will need O(n^2) time to find the median. In this method, we will find the highest element by traversing the array. We switch the last element with the highest element and traverse the array again to find the second-highest element. We keep doing this until we obtain the middle element/ the median.

```python
import time
import numpy as np

def median_using_passes(arr):
    n = len(arr)
    middle = n // 2
    
    for _ in range(n):
        max_idx = 0
        for i in range(1, n):
            if arr[i] > arr[max_idx]:
                max_idx = i
        
        arr[n - 1], arr[max_idx] = arr[max_idx], arr[n - 1]
        
        n -= 1 
        
        if middle >= n:
            return arr[middle]

np.random.seed(42)
arr = np.random.randint(1, 50, size=100)
print(arr)

start_time_passes = time.time()
result_passes = median_using_passes(arr.tolist())
end_time_passes = time.time()

print("Result using Passes method:", result_passes)

time_taken_passes = end_time_passes - start_time_passes
print("Time taken using Passes method:", time_taken_passes)
```

```python
def median_using_quickselect(arr):
    def partition(arr, low, high):
        pivot = arr[high]
        i = low - 1
        for j in range(low, high):
            if arr[j] <= pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]
        arr[i + 1], arr[high] = arr[high], arr[i + 1]
        return i + 1

    def quickselect(arr, low, high, k):
        if low < high:
            pivot_index = partition(arr, low, high)
            if pivot_index == k:
                return arr[pivot_index]
            elif pivot_index > k:
                return quickselect(arr, low, pivot_index - 1, k)
            else:
                return quickselect(arr, pivot_index + 1, high, k)
        else:
            return arr[low]

    n = len(arr)
    k = n // 2
    if n % 2 == 0:
        return (quickselect(arr, 0, n - 1, k - 1) + quickselect(arr, 0, n - 1, k)) / 2
    else:
        return quickselect(arr, 0, n - 1, k)

start_time_quickselect = time.time()
result_quickselect = median_using_quickselect(arr.tolist())
end_time_quickselect = time.time()

start_time_numpy = time.time()
result_numpy = np.median(arr)
end_time_numpy = time.time()

print("Result using QuickSelect method:", result_quickselect)
print("Result using NumPy's median method:", result_numpy)

time_taken_quickselect = end_time_quickselect - start_time_quickselect
time_taken_numpy = end_time_numpy - start_time_numpy

print("Time taken using QuickSelect method:", time_taken_quickselect)
print("Time taken using NumPy's median method:", time_taken_numpy)
```

# Question 5
## Use JAX to confirm the gradient evaluated by your method matches the analytical solution corresponding to a few random values of x and y

```python
import jax
import jax.numpy as jnp
import numpy as np

def f(x, y):
    return x*2 * y + y*3 * jnp.sin(x)

def analytical_gradient(x, y):
    df_dx = 2 * x * y + y**3 * jnp.cos(x)
    df_dy = x*2 + 3 * y*2 * jnp.sin(x)
    return df_dx, df_dy

np.random.seed(42)  
x_values = np.random.uniform(size=(3,))
y_values = np.random.uniform(size=(3,))

gradient_f = jax.grad(f)

for i in range(len(x_values)):
    x, y = x_values[i], y_values[i]
    analytical_grad = analytical_gradient(x, y)
    jax_grad = gradient_f(x, y)
    
    print("Data Points:", x, ",", y)
    print("Analytical Gradient:", analytical_grad)
    print("JAX Gradient:", jax_grad,"\n")
```

# Question 6
## Use sympy to confirm that you obtain the same gradient analytically.

```python
import sympy as sp

x, y = sp.symbols('x y')

f = x*2 * y + y*3 * sp.sin(x)

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("Gradient:")
print("df/dx =", df_dx)
print("df/dy =", df_dy)

numeric_gradient= sp.lambdify((x, y), (df_dx, df_dy), 'numpy')

x_val = 4.0
y_val = 5.0

analytical_grad = analytical_gradient(x_val, y_val)
numeric_grad = numeric_gradient(x_val, y_val)

print("Analytical Gradient:", analytical_grad)
print("Numeric Gradient:", numeric_grad)
```

# Question 8
## Store the same information using Python classes. We have an overall database which is a list of year objects. Each year contains a list of branches. Each branch contains a list of students. Each student has some properties like name, roll number and has marks in some subjects

```python
class records:
    def __init__(self):
        self.years = []

    def add_year(self, year):
        self.years.append(year)

class year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)

class student:
    def __init__(self, roll_number, name, marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks

class branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

def print_records(records):
    for year in records.years:
        print(f"Year: {year.year}")
        for branch in year.branches:
            print(f"  Branch: {branch.name}")
            for student in branch.students:
                print(f"    Roll Number: {student.roll_number}, Name: {student.name}")
                print("    Marks:")
                for subject, mark in student.marks.items():
                    print(f"      {subject}: {mark}")

# Creating instances for the given data
student_1 = student(1, 'N1', {'Maths': 100, 'English': 70})
student_2 = student(2, 'N2', {'Maths': 90, 'English': 80})

branch_1_2022 = branch('Branch 1')
branch_2_2022 = branch('Branch 2')
branch_1_2022.add_student(student_1)
branch_2_2022.add_student(student_2)

year_2022 = year(2022)
year_2022.add_branch(branch_1_2022)
year_2022.add_branch(branch_2_2022)

records = records()
records.add_year(year_2022)

print_records(records)
```

# Question 9
## Using matplotlib plot the following functions on the domain: x = 0.5 to 100.0 in steps of 0.5

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(f, x_min, x_max, step):
    
    x_values = np.arange(x_min, x_max + step, step)
    y_values = f(x_values)
    plt.plot(x_values, y_values)

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Plot of {f.__name__}')
    
    plt.show()
```

```python
def function_1(x):
    return x

plot_function(function_1, 0.5, 100, 0.5)
```

```python
def function_2(x):
    return x**2

plot_function(function_2, 0.5, 100, 0.5)
```

## Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as a, b, c, d, e. Find the column with the highest standard deviation. Find the row with the lowest mean.

```python
import numpy as np
import pandas as pd

matrix = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(matrix, columns=['a', 'b', 'c', 'd', 'e'])

max_std_col = df.std().idxmax()

min_mean_row = df.mean(axis=1).idxmin()

print(df)

print("\nColumn with the highest standard deviation:", max_std_col)

print("\nRow with the lowest mean:", min_mean_row)
```

## Write a small piece of code to explain broadcasting in numpy.

```python
import numpy as np

a = np.array([1.0, 2.0, 3.0])
b = np.array([2.0, 2.0, 2.0])
a * b
```

```python
a = np.array([1.0, 2.0, 3.0])
b = 2.0
a * b
```

Here, in the above codes, we see that in the second case, the size of arrays a and b are not the same, but still the multiplication operation is performed. This is possible because of broadcasting in Numpy. 

Broadcasting is possible if,

i) the dimensions of both the arrays are same

ii) the dimension of one array is 1

In the above case, the dimension of array b is one and hence it is broadcasted i.e. stretched so as to make the sizes of both arrays equal so that the multiplication operation can be performed.

```python
def argmin():
    input_str = input("Enter space-separated values for the array: ")
    arr = np.array([float(x) for x in input_str.split()])
        
    min_index = 0
    min_value = arr[0]

    for i in range(1, len(arr)):
        if arr[i] < min_value:
            min_value = arr[i]
            min_index = i

    numpy_argmin = np.argmin(arr)

    print("Argmin function:", arr[min_index])
    print("Numpy argmin:", arr[numpy_argmin])

argmin()
```
